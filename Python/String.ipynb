```
num = ["Hello", 34, 45, "", 40]

result = []

for item in num:
    if item != "":
        result.append(item)

print("Result:", result)
```

```
num = [2, 3, 4, 5, 2, 6, 3, 2]

result = []

for n in num:
    if n not in result:
        result.append(n)

print("Result:", result)
```

```
num = [2, 23, 24, 51, 46, 67]

even = []
odd = []

for n in num:
    if n % 2 == 0:
        even.append(n)
    else:
        odd.append(n)

print("Even", even)
print("Odd", odd)
```

```
char_list = ['P', 'y', 't', 'h', 'o', 'n']

reversed_list = []

for i in range(len(char_list) - 1, -1, -1):
    reversed_list.append(char_list[i])

print(reversed_list)
```
