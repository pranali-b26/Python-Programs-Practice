```
num = [10, 45, 23, 89, 12]

largest = max(num)
print("Largest Number:", largest)
```

```
num = [10, 23, 45, 60, 33]

result = []

for n in num:
    if n % 2 != 0:
        result.append(n)

print(result)
```

```
num = [1, 2, 3, 4, 5]

num.reverse()
print(num)
```

```
num = [2, 3, 4]

result = []

for n in num:
    result.append(n * n)

print(result)
```

```
num = ["Hello", 34, 45, "", 40]

result = []

for item in num:
    if item != "":
        result.append(item)

print("Result:", result)
```
