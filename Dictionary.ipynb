```python
d = {0:10, 1:20}
d[2] = 30
print(d)
```

```python
dic1 = {1:10, 2:20}
dic2 = {3:30, 4:40}
dic3 = {5:50, 6:60}

result = {}
result.update(dic1)
result.update(dic2)
result.update(dic3)

print(result)
```

```python
d = {1:10, 2:20, 3:30}
key = 2

print(key in d)
```

```python
d = {1:10, 2:20, 3:30}

for key, value in d.items():
    print(key, value)
```

```python
n = 5
d = {}

for x in range(1, n+1):
    d[x] = x*x

print(d)
```

```python
d = {}

for x in range(1,16):
    d[x] = x*x

print(d)
```

```python
d1 = {'a':1, 'b':2}
d2 = {'c':3, 'd':4}

d1.update(d2)
print(d1)
```

```python
d = {'a':100, 'b':200}

for k in d:
    print(k, d[k])
```

```python
d = {'a':10, 'b':20, 'c':30}
print(sum(d.values()))
```
