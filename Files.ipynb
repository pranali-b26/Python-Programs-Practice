```python
f1 = open("source.txt","w")
f1.write("Python File Handling Example")
f1.close()

f1 = open("source.txt","r")
f2 = open("destination.txt","w")

data = f1.read()
f2.write(data)

f1.close()
f2.close()

print("Source File Content:")
f1 = open("source.txt","r")
print(f1.read())
f1.close()

print("Destination File Content:")
f2 = open("destination.txt","r")
print(f2.read())
f2.close()
```

```python
file = open("numbers.txt","w")
file.write("10\n20\n30\n40")
file.close()

file = open("numbers.txt","r")

total = 0

for line in file:
    total = total + int(line)

print("Sum =",total)

file.close()
```

```python
file = open("words.txt","w")
file.write("The sun rises. The moon shines. the stars glow.")
file.close()

file = open("words.txt","r")

data = file.read().lower()

print("File Content:")
print(data)

count = data.count("the")

print("The appears",count,"times")

file.close()
```

```python
f1 = open("grades.txt","w")
f1.write("Alice A\nBob B\nCharlie C\nDavid A\nEva D")
f1.close()

f1 = open("grades.txt","r")
f2 = open("passed.txt","w")

for line in f1:
    name, grade = line.split()

    if grade == "A" or grade == "B":
        f2.write(name + "\n")

f1.close()
f2.close()

print("Passed Students:")

f2 = open("passed.txt","r")
print(f2.read())
f2.close()
```
