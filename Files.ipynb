```python
file = open("names.txt","w")

file.write("Alice\n")
file.write("Bob\n")
file.write("Charlie\n")

file.close()

file = open("names.txt","r")
print("names.txt Content:")
print(file.read())
file.close()
```

```python
file = open("data.txt","w")
file.write("Hello Python")
file.close()

file = open("data.txt","r")

data = file.read()

print("File Content:")
print(data)

print("Total characters:",len(data))

file.close()
```
